# Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read the data

```python
df = pd.read_csv('athlete_events.csv')
```

### Don't forget to drop duplicates first

```python
df = df.drop_duplicates()
```

# <font color='red'>**Beginner level**</font>

### Find the missing data in each column

```python
missing_values = df.isnull().sum()
print(missing_values)
```

### Number of males and females

```python
gender_counts = df['Sex'].value_counts()
print(gender_counts)
```

### How many cities do we have and what are they ?

```python
num_cities = df['City'].nunique()
city_names = df['City'].unique()
print('Number of cities:', num_cities)
print('City names:', city_names)
```

# <font color='red'>**Intermediate level**</fond>

### Find the number of athlete for each year

```python
athlete_counts = df.groupby('Year').size()
print(athlete_counts)
```

### Which athlete participated the most? (his/her name and ID)

```python
athlete_counts = df.groupby(['ID', 'Name']).size()
most_participated = athlete_counts.idxmax()
print('Most participated athlete:')
print('ID:', most_participated[0])
print('Name:', most_participated[1])
```

### What is the number of winners with the gold medal in each city?

```python
gold_df = df[df['Medal'] == 'Gold']
gold_counts = gold_df.groupby('City')['ID'].nunique()
print(gold_counts)
```

### Find the total missing data for each year?

```python
missing_counts = df.groupby('Year').apply(lambda x: x.isnull().sum())
print(missing_counts)
```
